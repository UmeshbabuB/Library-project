```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMassgeBox

root=Tk()
root.title('A Simple Crud Application!')

# Methods/Functions
def Database():
    global conn,cursor
    conn=sqlite3.connect('Book.db')
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS Book(book_title text, author text, year_of_publication integer, ISBN integer)')

    
def Add():
    if str(BOOK_TITLE.get(1.0,'end-1c'))=="" or str(AUTHOR.get(1.0,'end-1c'))=="" or str(YEAR_OF_PUBLICATION.get(1.0,'end-1c'))=="" or str(ISBN.get(1.0,'end-1c'))=="":
        txt_result.config(text="Please enter all the book details",fd="red")
    else:
        Database()
        cursor.execute('INSERT INTO Book(book_title text, author text, year_of_publication integer, ISBN integer) values(?,?,?,?)',(BOOK_TITLE.get(1.0,'end-1c'),AUTHOR.get(1.0,'end-1c'),YEAR_OF_PUBLICATION.get(1.0,'end-1c'),ISBN.get(1.0,'end-1c')))
        conn.commit()
        BOOK_TITLE.set("1.0","end")
        AUTHOR.set("1.0","end")
        YEAR_OF_PUBLICATION.set("1.0","end")
        ISBN.set("1.0","end")
        cursor.close()
        conn.close()
        txt_result.config(text="Record Inserted !",fg="green")
        
def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM Book ORDER BY book_title ASC")
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursor.close()
    txt_result.config(text="Successfully fetched data!",fg="black")

    
def Search():
    Database()
    tree.delete(*tree.get_children())
    if str(book_title.get(1.0, 'end-1c'))!="":
        cursor.execute(("select * from book_information where book_title = ")+"'"+str(book_title.get(1.0,'end-1c'))+"'")
        fetch=cursor.fetchall()
        
    elif str(author.get(1.0, 'end-1c'))!="":
        cursor.execute(("select * from book_information where author_name = ")+"'"+str(author_name.get(1.0,'end-1c'))+"'")
        fetch=cursor.fetchall()
        
    elif year_of_publication.get(1.0,'end-1c')!="":
        cursor.execute(("select * from book_information where year = ")+"'"+str(year.get(1.0,'end-1c'))+"'")
        fetch=cursor.fetchall()
    elif ISBN.get(1.0,'end-1c')!="":
        cursor.execute(("select * from book_information where ISBN = ")+"'"+str(ISBN.get(1.0,'end-1c'))+"'")
        fetch=cursor.fetchall()
    else:
        txt_result.config(text=" give atleast one identity!",fg="red")
        fetch=cursor.fetchall()
    for data in fetch:
        tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
        txt_result.config(text="please see the books!",fg="green")
    book_title.delete(1.0,'end-1c')
    author.delete(1.0,'end-1c')
    year_of_publication.delete(1.0,'end-1c')
    ISBN.delete(1.0,'end-1c')
    conn.commit()
    
    
def Issues():
    Database()
    selected_item=tree.selection()
    if len(selected_item)>0:
        tree.delete(selected_item)
        txt_result.config(text="book id issued!",fg='green')
    else:
        txt_result.config(text="please select atleast one book",fg='blue')
    
    book_title.delete(1.0,'end-1c')
    author.delete(1.0,'end-1c')
    year_of_publication.delete(1.0,'end-1c')
    ISBN.delete(1.0,'end-1c')
    conn.close()
    
    
def Exit():
    result=tkMassgeBox.askquestion("Do you want to exit? (y/n)",icon="warning")
    if result =="yes":
        root.destroy()
        exit()
        
def Delete():
    Database()
    if AUTHOR.get(1.0,'end-1c')!="":
        temp1="'"+str(AUTHOR.get(1.0,'end-1c'))+"'"
        temp="DELETE FROM Book WHERE author= "+temp1
        cursor.execute(temp)
        conn.commit()
        cursor.close()
    else:
        text_result.config(text="Please enter author!")
    
    
# Frame:-
Top=Frame(root,width=900,height=50,bd=8,relief='raise')
Top.pack(side=TOP)
Left=Frame(root,width=300,height=500,bd=8,relief='raise')
Left.pack(side=LEFT)
Right=Frame(root,width=600,height=500,bd=8,relief='raise')
Right.pack(side=RIGHT)

Forms=Frame(Left,width=300,height=450)
Forms.pack(side=TOP)
Buttons=Frame(Left,width=300,height=100,bd=10,relief='raise')
Buttons.pack(side=BOTTOM)


#Label Widget:-
txt_title=Label(Top,width=900,font=('arial',24),text="Python, Simple Crud Application")
txt_title.pack()
txt_book_title=Label(Forms,text="Book Title:",font=('arial',16),bd=15)
txt_book_title.grid(row=0,stick='e')
txt_author=Label(Forms,text="Author name:",font=('arial',16),bd=15)
txt_author.grid(row=1,stick='e')
txt_year_of_publication=Label(Forms,text="Year Of Publication:",font=('arial',16),bd=15)
txt_year_of_publication.grid(row=2,stick='e')
txt_ISBN=Label(Forms,text="ISBN:",font=('arial',16),bd=15)
txt_ISBN.grid(row=3,stick='e')
txt_result=Label(Buttons)
txt_result.pack(side=TOP)


# Text Widget:-
book_title=Text(Forms,height=2,width=30)
book_title.grid(row=0,column=1)
author_name=Text(Forms,height=2,width=30)
author_name.grid(row=1,column=1)
year=Text(Forms,height=2,width=30)
year.grid(row=2,column=1)
ISBN=Text(Forms,height=2,width=30)
ISBN.grid(row=3,column=1)


# Button Widget:-
btn_display=Button(Buttons,width=10,text="Display",command=Display)
btn_display.pack(side=LEFT)
btn_Search=Button(Buttons,width=10,text="Search",command=Search)
btn_Search.pack(side=LEFT)
btn_add=Button(Buttons,width=10,text="add",command=Add)
btn_add.pack(side=LEFT)
btn_issue=Button(Buttons,width=10,text="issue",command=Issues)
btn_issue.pack(side=LEFT)
btn_delete=Button(Buttons,width=10,text="delete",command=Delete)
btn_delete.pack(side=LEFT)
btn_exit=Button(Buttons,width=10,text="Exit",command=Exit)
btn_exit.pack(side=LEFT)


#LIST Widget:-
scrollbary=Scrollbar(Right,orient=VERTICAL)
scrollbarx=Scrollbar(Right,orient=HORIZONTAL)

tree=ttk.Treeview(Right,columns=("Book Title","Author","Year Of Publication","ISBN"))
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)


tree.heading("Book Title",text="Book Title",anchor=W)
tree.heading("Author",text="Author",anchor=W)
tree.heading("Year Of Publication",text="Year Of Publication",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)


tree.column("#0",minwidth=0,width=0)
tree.column("#1",minwidth=0,width=80)
tree.column("#2",minwidth=0,width=120)
tree.column("#3",minwidth=0,width=80)
tree.column("#4",minwidth=0,width=150)
tree.pack()


root.mainloop()
```
